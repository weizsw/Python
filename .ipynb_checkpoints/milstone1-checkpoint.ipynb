```python
from IPython.display import clear_output
import random

player1 = ''
player2 = ''
    
def player_input():
    player1 = ''
    player2 = ''
    while player1 != 'X' and player1 != 'O':
        player1 = input("Player1, Please pick a marker 'X' or 'O'\n")
    if player1 == 'X':
        player2 = 'O'
        print('\nPlayer1 now using X')
        print('Player2 now using O\n')
    elif player1 == 'O':
        player2 = 'X'
        print('\nPlayer1 now using O')
        print('Player2 now using X\n')
    return player1,player2

def place_marker(board, marker, position):
    board[position] = marker
    
    
def display_board(board):
    #clear_output()
    print(board[7]+'|'+board[8]+'|'+board[9])
    print(board[4]+'|'+board[5]+'|'+board[6])
    print(board[1]+'|'+board[2]+'|'+board[3]+'\n')
    
def win_check(board,mark):
    if mark == 'X' and mark == 'O':
        if board[7] == board[8] == board[9]:
            return True
        elif board[4] == board[5] == board[6]:
            return True
        elif board[1] == board[2] == board[3]:
            return True
        elif board[7] == board[4] == board[1]:
            return True
        elif board[8] == board[5] == board[2]:
            return True
        elif board[9] == board[6] == board[3]:
            return True
        elif board[7] == board[5] == board[3]:
            return True
        elif board[9] == board[5] == board[1]:
            return True
        else:
            return False
    else:
        return False
    
def go_first():
    
    dice = random.randint(1,100)
    
    if dice %2 == 0:
        return 1
    else:
        return 2
    
def space_check(board,position):
    if board[position] != ' ':
        return False
    else:
        return True
    
def full_board_check(board):
    for item in board:
        if item != ' ':
            return False
        else:
            return True
        
def player_choice(board):
    x = int(input('Where you wanna place the marker? 1-9\n'))
    if space_check(board,x) == True:
        return x
    else:
        print('The place has been taken, please choose another\n')
        full_board_check(board)  
        
def replay():
    decision = input('Do you wanna play again? (Yes/No)\n')
    if decision == 'Yes':
        return True
    elif decision == 'No':
        return False
```

```python
print('Welcome to Tic Tac Toe!')
board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
while True:
    player1, player2 = player_input()
    display_board(board)
    player_now = go_first
    if player_now == 1:
        print('Player1 goes first!')
        current_marker = player2
    else:
        print('Player2 goes first!')
        current_marker = player1
    while win_check != False:
        if current_marker == player2:
            current_marker = player1
            player_position = player_choice(board)
            if space_check(board,player_position) != False:
                place_marker(board,current_marker,player_position)
                display_board(board)
            else:
                print('Space has been taken')
        elif current_marker == player1:
            current_marker = player2
            player_position = player_choice(board)
            if space_check(board,player_position) != False:
                place_marker(board,current_marker,player_position)
                display_board(board)
            else:
                print('Space has been taken')
```
